```python
import matplotlib.pyplot as plt

# Dados
labels = ['Neo Legibility', 'TF-IDF', 'Media']
sizes = [15, 10, 8]  # Porcentagem de cada categoria
colors = ['orange', 'lightblue', 'red']  # Cores das fatias

# Criar o gráfico de pizza
plt.figure(figsize=(6,6))  # Define o tamanho da figura
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140, textprops={'fontsize': 12})

# Adicionar título
#plt.title('Contagem do Melhor Modelo')

# Mostrar o gráfico
plt.axis('equal')  # Assegura que o gráfico de pizza seja desenhado como um círculo
plt.show()
```
