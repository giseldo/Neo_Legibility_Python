```python
import matplotlib.pyplot as plt
import pandas as pd

dados = [['Neo Legibility',	15, 'orange'],
         ['TF-IDF', 10, 'blue'],
         ['bow', 4, 'green'],
         ['media', 4, 'red']]

colunas = ['modelo', 'valor', 'cor']
df = pd.DataFrame(data=dados, columns=colunas)
df
```

```python
plt.bar(df["modelo"], df["valor"], color=df["cor"])
plt.title("Consolidado dos 34 Projetos")
plt.ylabel("Quantidade de vezes em que o modelo teve menor MAE")
plt.xlabel("Modelos")
plt.show()
```

```python
# intervalo de confiança
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Group': ['Usability', 'Ease of use', 'External variables', 'attitude'],
    'Cronbach': [0.81, 0.92, 0.87, 0.73],
    'CI_lower': [0.64, 0.87, 0.77, 0.49],
    'CI_upper': [0.90, 0.95, 0.93, 0.85]
}

df = pd.DataFrame(data)

plt.errorbar(df['Group'], df['Cronbach'], yerr=[df['Cronbach'] - df['CI_lower'], df['CI_upper'] - df['Cronbach']], capsize=5, ecolor='black', fmt='o')

plt.xlabel('Variables')
plt.ylabel('Cronbach')
plt.title('Cronbach of TAM Variables')
plt.grid(True)

plt.show()
```
